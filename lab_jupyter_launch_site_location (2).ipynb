```python
# TASK 3: Calculate the distances between a launch site to its proximities


import pandas as pd
from math import radians, cos, sin, sqrt, atan2

# Define the Haversine formula to calculate distances
def haversine(lat1, lon1, lat2, lon2):
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    
    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    r = 6371  # Radius of Earth in kilometers. Use 3956 for miles
    return r * c

# Sample nearby landmarks/cities with their coordinates
nearby_points = {
    'City': ['Orlando', 'Miami', 'Los Angeles', 'San Francisco'],
    'Latitude': [28.538336, 25.761680, 34.052235, 37.774929],
    'Longitude': [-81.379234, -80.191790, -118.243683, -122.419418]
}

# Convert to DataFrame
nearby_df = pd.DataFrame(nearby_points)

# Display the launch sites DataFrame
print("Launch Sites DataFrame:")
print(launch_sites_df)

# Calculate distances from each launch site to each nearby point
distances = []

for index, site in launch_sites_df.iterrows():
    for _, point in nearby_df.iterrows():
        distance = haversine(site['Lat'], site['Long'], point['Latitude'], point['Longitude'])
        distances.append({
            'Launch Site': site['Launch Site'],
            'Proximity Point': point['City'],
            'Distance (km)': distance
        })

# Convert the distances to a DataFrame
distances_df = pd.DataFrame(distances)

# Display the distances DataFrame
print("\nDistances from Launch Sites to Proximity Points:")
print(distances_df)
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site

import folium
from folium.features import DivIcon
from folium.plugins import MousePosition

# Function to calculate distance using Haversine formula
def haversine(lat1, lon1, lat2, lon2):
    from math import radians, cos, sin, sqrt, atan2
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    r = 6371  # Radius of Earth in kilometers. Use 3956 for miles
    return r * c

# Example data: Launch site coordinates
launch_sites_data = {
    'Launch Site': ['CCAFS LC-40'],
    'Lat': [28.562302],
    'Long': [-80.577356]
}

# Convert to DataFrame
launch_sites_df = pd.DataFrame(launch_sites_data)

# Use one of the launch sites as an example
launch_site = launch_sites_df.iloc[0]  # Example: CCAFS LC-40

# Create a Folium map centered at the selected launch site
launch_site_lat = launch_site['Lat']
launch_site_lon = launch_site['Long']
spacex_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=12)

# Add a marker for the launch site
folium.Marker(
    location=[launch_site_lat, launch_site_lon],
    popup=f"{launch_site['Launch Site']} (Launch Site)",
    icon=folium.Icon(color='blue', icon='info-sign')
).add_to(spacex_map)

# Define nearby points of interest
points_of_interest = [
    {"name": "Closest Coastline", "lat": 28.56367, "lon": -80.57163},
    {"name": "Closest City (Orlando)", "lat": 28.538336, "lon": -81.379234},
    {"name": "Closest Railway", "lat": 28.5721, "lon": -80.5853},
    {"name": "Closest Highway", "lat": 28.5634, "lon": -80.5708}
]

# Add MousePosition plugin to get coordinates
MousePosition().add_to(spacex_map)

# Calculate distances and add markers and lines
for point in points_of_interest:
    # Calculate distance
    distance = haversine(launch_site_lat, launch_site_lon, point["lat"], point["lon"])
    
    # Add marker with distance
    folium.Marker(
        location=[point["lat"], point["lon"]],
        popup=f"{point['name']}<br>Distance: {distance:.2f} km",
        icon=DivIcon(
            icon_size=(150, 36),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12; color:#d35400;"><b>{point["name"]} ({distance:.2f} KM)</b></div>'
        )
    ).add_to(spacex_map)
    
    # Draw line from launch site to the point
    line = folium.PolyLine(
        locations=[(launch_site_lat, launch_site_lon), (point["lat"], point["lon"])],
        weight=1,
        color='red'
    )
    spacex_map.add_child(line)

# Display the map
spacex_map
```
